```python
viviendas =['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Flat',
 'casa de villa',
 'Cochera',
 'Loft',
 'Chalet',
 ]
```

```python
viviendas
```

```python
info_viviendas=info_alquiler['Tipo'].isin(viviendas)
```

```python
info_viviendas
```

```python
datos_viviendas=info_alquiler[info_viviendas]
```

```python
datos_viviendas
```

```python
datos_viviendas.index=range(datos_viviendas.shape[0])
```

```python
datos_viviendas
```

```python
datos_viviendas.head(30)
```

```python
sel_departamento=datos_viviendas['Tipo']=="Departamento"
cant_departamento=datos_viviendas[sel_departamento].shape[0]
cant_departamento
```

```python
sel_otros_departamento=datos_viviendas['Tipo']=="Departamento"
cant_otras=datos_viviendas[~sel_otros_departamento].shape[0]
cant_otras
```

```python
sel_casa_hab_coch=(datos_viviendas['Tipo']=="Casa") | (datos_viviendas['Tipo']=="Habitacion") | (datos_viviendas['Tipo']=="Cochera")
cant_casa_hab_coch=datos_viviendas[sel_casa_hab_coch].shape[0]
cant_casa_hab_coch
```

```python
sel_area=(datos_viviendas['Area']>=60) & (datos_viviendas['Area']<=100)
cant_viv_area=datos_viviendas[sel_area]
cant_viv_area
```

```python
sel_hab_precio=(datos_viviendas['Cuartos']>=4) & (datos_viviendas['Valor']<=2000)
cant_hab_precio=datos_viviendas[sel_hab_precio].shape[0]
cant_hab_precio
```

```python
print("Se encontraron los sgtes resultados:")
print(f"La cantidad de departamentos es de: {cant_departamento}")
print(f"La cantidad de otras viviendas: {cant_otras}")
print(f"La cantidad de viviendas tipo casa, habitacion y cochera es de: {cant_casa_hab_coch}")
print(f"La cantidad de viviendas con un area de 60 y 100 metros cuadrados es de :{cant_viv_area}")
print(f"La cantidad de viviendas con mas de 4 habitaciones y un valor menor o igual a 2000 reales es de: {cant_hab_precio}")
```
